Нормализуем данные

```python
nlp = spacy.load('en_core_web_sm')

def normalize(tweet):
  tweets = " ".join(filter(lambda x: x[0]!= '@' , tweet.split()))
  tweets = " ".join(filter(lambda x: x[0]!= '#' , tweets.split()))
  tweets = " ".join(filter(lambda x: x not in stopwords.words('english') , tweets.split()))
  tweets = re.sub('[^a-zA-Z]', ' ', tweets)
  tweets = tweets.lower()
  tweets = nlp(tweets)
  tweets = " ".join([token.lemma_ for token in tweets])
  tweets = re.sub(' +', ' ', tweets)
  tweets = " ".join(filter(lambda x: x not in ['PRON', '-PRON-'] , tweets.split()))
  return tweets
```

```python
tqdm.pandas()

data['normalized_text'] = data.raw_text.progress_apply(normalize)
data.head()
```

```python
import re

data['tags'] = data.raw_text.apply(lambda x: ' '.join(re.findall(r"#(\w+)", x)))
data['mentions'] = data.raw_text.apply(lambda x: ' '.join(re.findall(r"@(\w+)", x)))
data.head()
```

### Кодируем твиты

Кодируем твиты при помощи Universal Sentence Encoder и TF-IDF

```python
Y = pd.get_dummies(data.label).values
Y = np.argmax(Y, axis=1)
```

```python
nlp = spacy.load('en_use_md')

def vectorize(tweet):
  doc = nlp(tweet)
  return doc.vector

data['tweet_vector_use'] = data.normalized_text.progress_apply(vectorize)
```

```python
X_use = data.tweet_vector_use
X_use = np.array(X_use)
```

```python
from sklearn.model_selection import train_test_split

X_train_use, X_test_use, y_train_use, y_test_use = train_test_split(
    X_use, 
    Y, 
    test_size=0.25, 
    random_state=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


X_train_tf, X_test_tf, y_train_tf, y_test_tf = train_test_split(
    data.normalized_text, 
    Y, 
    test_size=0.25, 
    random_state=1)



tfidf = TfidfVectorizer(stop_words=stopwords.words('english'))
tfidf.fit(X_train_tf.to_list())
```

```python
X_train_tf = tfidf.transform(X_train_tf)
X_test_tf = tfidf.transform(X_test_tf)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, accuracy_score, f1_score

def fitClassifiers(classifiers, X_train, X_test, y_train, y_test):
    f1_scores = {}
    accuracy_scores = {}
    
    for classifier in classifiers:
        model_name = type(classifier).__name__
        
        classifier.fit(X_train, y_train)
        probs = classifier.predict_proba(X_test)
        pred = classifier.predict(X_test)

        accuracy = accuracy_score(y_test, pred)
        f1 = f1_score(y_test, pred, average='macro')

        f1_scores[model_name] = f1
        accuracy_scores[model_name] = accuracy
    
    return classifiers, accuracy_scores, f1_scores
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier
from lightgbm import LGBMClassifier

classifiers = [
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    ExtraTreesClassifier(),
    LGBMClassifier(),
    AdaBoostClassifier()
]
```

```python
X_train_sk = pd.DataFrame(X_train_use.tolist(), index=range(X_train_use.shape[0]))
X_test_sk = pd.DataFrame(X_test_use.tolist(), index=range(X_test_use.shape[0]))
```

```python
classifiers, accuracy_scores_use, f1_scores_use = fitClassifiers(
    classifiers, 
    X_train_sk, 
    X_test_sk,
    y_train_use,
    y_test_use
    )
```

```python
classifiers = [
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    ExtraTreesClassifier(),
    LGBMClassifier(),
    AdaBoostClassifier()
]

classifiers, accuracy_scores_tfidf, f1_scores_tfidf = fitClassifiers(
    classifiers, 
    X_train_tf, 
    X_test_tf,
    y_train_tf,
    y_test_tf
    )
```

### Попробуем обучить полносвязную нейронную сеть

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    # tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    # tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(3)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
X_train_tensor = pd.Series(X_train_use).apply(tf.convert_to_tensor).to_list()
X_train_tensor = tf.convert_to_tensor(X_train_tensor)

X_test_tensor = pd.Series(X_test_use).apply(tf.convert_to_tensor).to_list()
X_test_tensor = tf.convert_to_tensor(X_test_tensor)
```

```python
model.fit(X_train_tensor, y_train_use, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(X_test_tensor,  y_test_use, verbose=2)

print('\nTest accuracy:', test_acc)
```
